```python
import matplotlib.pyplot as plt
import torchvision
import torchvision.transforms as transforms

# Load CIFAR-10 dataset
transform = transforms.Compose([
    transforms.ToTensor()  # Convert PIL images to tensors
])

dataset = torchvision.datasets.CIFAR10(root="D:/KMUTT/Year 4th/ENE490 DL/Coding/Dataset/RN/train", train=True, download=False, transform=transform)

# Access a single image and label
image, label = dataset[0]
print(f"Image shape: {image.shape}, Label: {label}")  # Tensor shape and label index

# Visualize the first image in the dataset
plt.imshow(image.permute(1, 2, 0))  # Rearrange tensor dimensions from (C, H, W) to (H, W, C)
plt.title(f"Label: {dataset.classes[label]}")  # Show the class name
plt.show()
```

```python
from torch.utils.data import DataLoader

# Create a DataLoader
dataloader = DataLoader(dataset, batch_size=4, shuffle=True)

# Get a batch of images and labels
data_iter = iter(dataloader)
images, labels = next(data_iter)

# Print the shape of the batch
print(f"Batch of images shape: {images.shape}")  # Shape: (batch_size, channels, height, width)

# Visualize the batch
fig, axes = plt.subplots(1, 4, figsize=(12, 3))
for i in range(4):
    axes[i].imshow(images[i].permute(1, 2, 0))  # Rearrange dimensions
    axes[i].set_title(f"Label: {dataset.classes[labels[i]]}")
    axes[i].axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import torchvision
import torchvision.transforms as transforms

# Define transformation to increase resolution
transform = transforms.Compose([
    transforms.Resize((512, 512)),  # Resize images to 224x224
    transforms.ToTensor()
])

# Load CIFAR-10 dataset with increased resolution
dataset = torchvision.datasets.CIFAR10(root="D:/KMUTT/Year 4th/ENE490 DL/Coding/Dataset/RN/train", train=True, download=False, transform=transform)

# Visualize a single image
image, label = dataset[0]
print(f"Image shape after resizing: {image.shape}, Label: {label}")  # Tensor shape and label

# Display the resized image
plt.imshow(image.permute(1, 2, 0))
plt.title(f"Label: {dataset.classes[label]}")  # Show class name
plt.show()
```

```python
from torch.utils.data import DataLoader

# Create DataLoader
dataloader = DataLoader(dataset, batch_size=4, shuffle=True)

# Get a batch of images and labels
data_iter = iter(dataloader)
images, labels = next(data_iter)

# Visualize resized images
fig, axes = plt.subplots(1, 4, figsize=(12, 3))  # Increased figsize for clarity
for i in range(4):
    axes[i].imshow(images[i].permute(1, 2, 0))  # Rearrange dimensions (C, H, W) -> (H, W, C)
    axes[i].set_title(f"Label: {dataset.classes[labels[i]]}")
    axes[i].axis('off')
plt.show()
```
